```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
seaborn.set_style("white")

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pandas.read_csv(url, header = None, sep = ",")
wine
```

```python
names = ["class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
"Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
names
```

```python
wine.columns = names
wine
```

```python
wine.drop(columns="class").corr().round(1)
```

```python
wine_bis = scale(wine.drop(columns = "class"))
pca = PCA(n_components = 13)
pca.fit(wine_bis)
```

```python
eig = pandas.DataFrame(
    {
        "Dimension" : ["Dim" + str(x + 1) for x in range(13)], 
        "Valeur propre" : pca.explained_variance_,
        "% variance expliquÃ©e" : numpy.round(pca.explained_variance_ratio_ * 100),
        "% cum. var. expliquÃ©e" : numpy.round(numpy.cumsum(pca.explained_variance_ratio_) * 100)
    },
    columns = ["Dimension", "Valeur propre", "% variance expliquÃ©e", "% cum. var. expliquÃ©e"]
)
eig
```

```python
plt.figure(figsize=(16, 6))
g_eig = seaborn.barplot(x = "Dimension", 
                        y = "% variance expliquÃ©e",
                        palette = ["lightseagreen"],
                        data = eig)
plt.text(3.25, 26, "25%")
plt.axhline(y = 100/13, linewidth = .5, color = "dimgray", linestyle = "--")
g_eig.set(ylabel = "Variance expliquÃ©e (%)")
g_eig.figure.suptitle("Variance expliquÃ©e par dimension")

plt.show()
```

```python
wine_pca = pca.transform(wine_bis)
wine_pca_df = pandas.DataFrame({
    "Dim1" : wine_pca[:,0], 
    "Dim2" : wine_pca[:,1], 
    "Dim3" : wine_pca[:,2], 
    "Class" : wine["class"]
})
wine_pca_df.head()
```

```python
coordvar = pca.components_.T * numpy.sqrt(pca.explained_variance_)
coordvar_df = pandas.DataFrame(coordvar, columns=['PC'+str(i+1) for i in range(13)], index=names[1:])
coordvar_df
```

```python
fig, axes = plt.subplots(figsize = (10, 10))
fig.suptitle("Cercle des corrÃ©lations")
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)
axes.axvline(x = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
axes.axhline(y = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
for j in range(13):
    axes.text(coordvar_df["PC1"][j],coordvar_df["PC2"][j], coordvar_df.index[j], size = 25)
    axes.plot([0,coordvar_df["PC1"][j]], [0,coordvar_df["PC2"][j]], color = "gray", linestyle = 'dashed')
plt.gca().add_artist(plt.Circle((0,0),1,color='blue',fill=False))

plt.show()
```

```python
g_pca = seaborn.lmplot(x = "Dim1", y = "Dim2", hue = "Class", data = wine_pca_df, fit_reg = False, 
                       height = 4, aspect = 3)
g_pca.set(xlabel = "Dimension 1 (73%)", ylabel = "Dimension 2 (23 %)")
g_pca.fig.suptitle("Premier plan factoriel")

plt.show()
```
